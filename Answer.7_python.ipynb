```python
import pandas as pd
import matplotlib.pyplot as plt
import requests

# Define the URL
url = "https://data.nasa.gov/resource/y77d-th95.json"

# Send a GET request to the URL and retrieve the data
response = requests.get(url)

# Load the data into a pandas DataFrame
data = pd.DataFrame(response.json())

# Convert the year column to datetime format
data['year'] = pd.to_datetime(data['year'])

# Filter the DataFrame to get Earth meteorites that fell before the year 2000
earth_meteorites_before_2000 = data[(data['reclat'] != 'Unknown') & (data['year'].dt.year < 2000)]

# Draw a scatter plot to visualize the Earth meteorites before the year 2000
plt.scatter(earth_meteorites_before_2000['reclong'], earth_meteorites_before_2000['reclat'])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Earth Meteorites Before 2000')
plt.show()

# Print the filtered DataFrame
print(earth_meteorites_before_2000)
```
