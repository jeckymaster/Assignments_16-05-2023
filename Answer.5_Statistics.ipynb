A pharmaceutical company is developing a drug that is supposed to reduce blood
pressure. They conduct a clinical trial with 100 patients and record their blood
pressure before and after taking the drug. The company wants to know if the change
in blood pressure follows a normal distribution.
https://drive.google.com/file/d/1mCjtYHiX--mMUjicuaP2gH3k-SnFxt8Y/view?usp=share_

```python
import requests as req
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/jeckymaster/Assignments_16-05-2023/main/data_Blood%2BPressure.csv"
df = pd.read_csv(url)
```

```python
df.columns
```

```python
df['difference'] = df[' Blood Pressure Before (mmHg)'] - df[' Blood Pressure After (mmHg)']
```

```python
df.head()
```

```python
plt.hist(df['difference'], bins="auto", edgecolor='black')
plt.xlabel('Change in Blood Pressure (mmHg)')
plt.ylabel('Frequency')
plt.title('Distribution of Change in Blood Pressure')
plt.show()
```

```python
import scipy.stats as stats
stats.probplot(df['difference'], dist="norm", plot=plt)
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Ordered Values')
plt.title('Q-Q Plot')
plt.show()
```

```python
z_scores = stats.zscore(df['difference'])
```

```python
extreme_values = [score for score in z_scores if abs(score) > 3] #

if len(extreme_values) == 0:
    print("The distribution appears to be approximately normal.")
else:
    print("The distribution does not appear to be normal.")
```
