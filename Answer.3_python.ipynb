# The proper code with function

```python
import requests as req
import pandas as pd
import numpy as np

def download_pokemon(url):
    resp = req.get(url)
    json_data = resp.json()
    data_pokemon = json_data["pokemon"]
    return pd.DataFrame(data_pokemon)

def save_2_excel(df, filename):
    df.to_excel(file_name, index=False)
    print(f"The Dataframe saved successfully as {filename}")
```

```python
url = "https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json"
df_poke = download_pokemon(url)
```

```python
df_poke.head()
```

```python
df_poke.info()
```

```python
df_poke['candy_count'] = df_poke['candy_count'].fillna(0)  # Handling Nan value
df_poke['egg'] = df_poke['egg'].replace('Not in Eggs', 0)  # replacing string to 0
df_poke['egg'] = df_poke['egg'].replace('Omanyte Candy', 0) # replacing string to 0
```

```python
df_poke['egg'] = df_poke['egg'].fillna(0)  # Handling Nan value
```

```python
df_poke['num'] = df_poke['num'].astype(int)
df_poke['height'] = df_poke['height'].str.replace(' m', '').astype(float)
```

```python
df_poke['weight'] = df_poke['weight'].str.replace(' kg', '').astype(float)
```

```python
df_poke['candy_count'] = df_poke['candy_count'].astype(int)
```

```python
df_poke['egg'] = df_poke['egg'].str.replace(' km', '').astype(float)
```

```python
filename = "pokemon_pokedex.xlsx"
save_2_excel(df_poke, filename)
```
